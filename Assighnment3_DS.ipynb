# Introduction

Descriptive Analytics for Numerical Columns
●	Objective: To compute and analyze basic statistical measures for numerical columns in the dataset.
●**	Steps:**
	Load the dataset into a data analysis tool or programming environment (e.g., Python with pandas library).
	Identify numerical columns in the dataset.
	Calculate the mean, median, mode, and standard deviation for these columns.
	Provide a brief interpretation of these statistics.


```python
import pandas as pd
import scipy.stats as stats
```

```python
df=pd.read_csv("/content/sales_data_with_discounts.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
#STATS FOR VOLUME COLOUMN
vol_mean=df['Volume'].mean()
vol_median=df['Volume'].median()
vol_mode=df['Volume'].mode().iloc[0]
vol_std=df['Volume'].std()
print("Mean of volume=",vol_mean)
print("Median of volume=",vol_median)
print("Mode of volume=",vol_mode)
print("StandardDeviation of volume=",vol_std)
```

```python
#STATS FOR AVG PRICE COLOUMN
avg_mean=df['Avg Price'].mean()
avg_median=df['Avg Price'].median()
avg_mode=df['Avg Price'].mode().iloc[0]
avg_std=df['Avg Price'].std()
print("Mean of AvgPrice=",avg_mean)
print("Median of AvgPrice=",avg_median)
print("Mode of AvgPrice=",avg_mode)
print("StandardDeviation of AvgPrice=",avg_std)
```

```python
# Stats for Total Sales Value column
total_sales_mean = df['Total Sales Value'].mean()
total_sales_median = df['Total Sales Value'].median()
total_sales_mode = df['Total Sales Value'].mode().iloc[0]
total_sales_std = df['Total Sales Value'].std()

print("Mean of Total Sales Value =", total_sales_mean)
print("Median of Total Sales Value =", total_sales_median)
print("Mode of Total Sales Value =", total_sales_mode)
print("Standard Deviation of Total Sales Value =", total_sales_std)
```

```python
# Stats for Discount Rate (%) column
discount_rate_mean = df['Discount Rate (%)'].mean()
discount_rate_median = df['Discount Rate (%)'].median()
discount_rate_mode = df['Discount Rate (%)'].mode().iloc[0]
discount_rate_std = df['Discount Rate (%)'].std()

print("Mean of Discount Rate (%) =", discount_rate_mean)
print("Median of Discount Rate (%) =", discount_rate_median)
print("Mode of Discount Rate (%) =", discount_rate_mode)
print("Standard Deviation of Discount Rate (%) =", discount_rate_std)
print()
```

```python
# Stats for Discount Amount column
discount_amount_mean = df['Discount Amount'].mean()
discount_amount_median = df['Discount Amount'].median()
discount_amount_mode = df['Discount Amount'].mode().iloc[0]
discount_amount_std = df['Discount Amount'].std()

print("Mean of Discount Amount =", discount_amount_mean)
print("Median of Discount Amount =", discount_amount_median)
print("Mode of Discount Amount =", discount_amount_mode)
print("Standard Deviation of Discount Amount =", discount_amount_std)
print()
```

```python
# Stats for Net Sales Value column
net_sales_mean = df['Net Sales Value'].mean()
net_sales_median = df['Net Sales Value'].median()
net_sales_mode = df['Net Sales Value'].mode().iloc[0]
net_sales_std = df['Net Sales Value'].std()

print("Mean of Net Sales Value =", net_sales_mean)
print("Median of Net Sales Value =", net_sales_median)
print("Mode of Net Sales Value =", net_sales_mode)
print("Standard Deviation of Net Sales Value =", net_sales_std)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
categorical_cols = df.select_dtypes(include=['object'])
categorical_cols
```

```python
# Histogram for Volume
plt.figure(figsize=(6,4))
sns.histplot(data=df, x='Volume', kde=True)
plt.title('Histogram of Volume')
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.show()

# Boxplot for Volume
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Volume')
plt.title('Boxplot of Volume')
plt.xlabel('Volume')
plt.show()
```
