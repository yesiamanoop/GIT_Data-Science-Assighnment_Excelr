# BASIC STATISTICS-2

Context:
A hospital maintains patient records including admission details, department, diagnosis, doctor, and bill amount. You have two datasets: one with patient info and another with billing details. Some patients have blank bill amounts, and there are multiple rows for the same patient due to follow-ups.


Tasks:
1.	Load the patient dataset and show summary with info().

2.	Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].

3.	Drop administrative columns like ['ReceptionistID', 'CheckInTime'].

4.	Use groupby to find total bill amount per department.

5.	Remove duplicate patient records based on PatientID.

6.	Fill missing BillAmount values with the mean bill amount.

7.	Merge the billing dataset with patient dataset on PatientID.

8.	Concatenate an additional DataFrame that contains new patients for the current week (row-wise).

9.	Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).


Expected Outcome:

•	Final cleaned dataset with accurate billing info.

•	All missing values handled, merged dataset across PatientID.

•	Ability to perform further analytics on department-wise revenue or doctor performance.


```python
import pandas as pd
```

```python
pdf=pd.read_csv("/content/Patient_Data.csv")
pdf
```

```python
pdf.info()
```

```python
pdf.drop(['ReceptionistID', 'CheckInTime'], axis=1,inplace=True)
```

```python
pdf
```

```python
pdf.groupby("Department")["BillAmount"].sum()
```

```python
pdf.duplicated("PatientID").sum()
```

```python
spdf=pdf.drop_duplicates("PatientID")
```

```python
spdf.duplicated().sum()
```

```python
spdf.isnull().sum()
```

```python

mean_value = spdf["BillAmount"].mean()
spdf["BillAmount"] = spdf["BillAmount"].fillna(mean_value)
```

```python
spdf
```

```python
bdf = pd.read_csv('Billing_Data.csv')
merged_df = pd.merge(spdf, bdf, on="PatientID", how="inner")
```

```python
merged_df
```

```python

spdf = {'PatientID': [105, 106], 'Name': ['Anoop', 'Mohan'], 'Department': ['Cardiology', 'Dermatology'],'Doctor': ['Dr. Rose', 'Dr. Rose']}
new_patients_df = pd.DataFrame(spdf)


updated_patient_df = pd.concat([merged_df,new_patients_df], ignore_index=True)
updated_patient_df
```

```python
mean_value = updated_patient_df["BillAmount"].mean()
updated_patient_df["BillAmount"] = updated_patient_df["BillAmount"].fillna(mean_value)


mean_value = updated_patient_df["InsuranceCovered"].mean()
updated_patient_df["InsuranceCovered"] = updated_patient_df["InsuranceCovered"].fillna(mean_value)

mean_value = updated_patient_df["FinalAmount"].mean()
updated_patient_df["FinalAmount"] = updated_patient_df["FinalAmount"].fillna(mean_value)

```
